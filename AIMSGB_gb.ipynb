```python
from aimsgb import GrainBoundary, Grain
from pymatgen.core import Structure

base_path = "/Users/emilydai/Library/CloudStorage/OneDrive-///AI/Random"
save_path = base_path + "/LATP_GB.vasp"

input_structure = Grain.from_file('/Users/emilydai/Library/CloudStorage/OneDrive-//Ncku lab/ML_LATP/poscar/LATP_2Al/POSCAR')
gb = GrainBoundary([0, 0, 1], 13, [2, 3, 0], input_structure)
structure = Grain.stack_grains(gb.grain_a, gb.grain_b, direction=gb.direction)

structure = structure.get_sorted_structure()
structure.to(fmt="poscar", filename=save_path)
```
